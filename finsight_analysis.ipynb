```python
import pandas as pd
```

```python
df = pd.read_csv("store_customers.csv")
```

```python
df.rename(columns={
    "Annual Income (k$)": "Income",
    "Spending Score (1-100)": "SpendingScore"
}, inplace=True)
```

```python
def segment(row):
    if row["Income"] > 70 and row["SpendingScore"] > 60:
        return "Premium"
    elif row["Income"] > 70 and row["SpendingScore"] <= 60:
        return "Potential"
    elif row["Income"] <= 70 and row["SpendingScore"] > 60:
        return "Impulsive"
    else:
        return "Budget"
df["Segment"] = df.apply(segment, axis=1)
```

```python
def churn(row):
    if row["SpendingScore"] < 30 and row["Income"] > 60:
        return "High Risk"
    elif row["SpendingScore"] < 50:
        return "Medium Risk"
    else:
        return "Low Risk"

df["ChurnRisk"] = df.apply(churn, axis=1)
```

```python
def opportunity(row):
    if row["Income"] > 70 and row["SpendingScore"] < 40:
        return "High Opportunity"
    elif row["Income"] > 50 and row["SpendingScore"] < 60:
        return "Medium Opportunity"
    else:
        return "Low Opportunity"

df["RevenueOpportunity"] = df.apply(opportunity, axis=1)
```

```python
df["FinancialHealthScore"] = (
    (df["SpendingScore"] * 0.6) + (df["Income"] * 0.4)
).round(1)

# Save file
df.to_csv("finSight_data.csv", index=False)

print("Done! File saved as finSight_data.csv")
```
